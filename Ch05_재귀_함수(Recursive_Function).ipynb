## DFS/BFS 사전 지식
## 재귀 함수(Recursive Function)
* DFS와 BFS를 구현하려면 재귀 함수도 이해하고 있어야 한다.
* <code>재귀 함수(Recursive Function)</code>이란 자기 자신을 다시 호출하는 함수를 의미한다. 


가장 간단한 재귀함수부터 알아보자!

```
# 5-3.py 재귀 함수 예제
def recursive_function():
    print("재귀 함수를 호출합니다.")
    recursive_function()  # 자기 자신을 다시 호출한다

recursive_function()
```

다음과 같은 에러 메세지를 볼 수 있다.<br>
"RecursionError: maximum recursion depth exceeded in comparison" <br>

최대 재귀(Recursive) 깊이를 초과했다는 내용이다. 보통 파이썬 인터프리터는 호출 횟수 제한이 있는데 이 한계를 벗어났기 떄문이다. 따라서 재귀 함수를 사용할 때는 무한대로 재귀 호출을 진행할 수 없다는 점을 기억하

### 재귀 함수의 종료 조건

* 재귀 함수를 문제 풀이에서 사용할 때는 재귀 함수가 언제 끝날지, 종료 조건을 반드시 명시해야 한다
* 종료 조건을 명시하지 않으면 함수가 무한히 호출될 수 있다

재귀 함수를 100번 호출하도록 코드를 작성해보자 

```
def recursive_function(i):
    # 100번째 출력했을 때 종료되도록 종료 조건 명시
    if i == 100:
        return
    print(i, "번째 재귀 함수에서", i + 1, "번째 재귀 함수를 호출합니다.")
    recursive_function(i+1)
    print(i, "번째 재귀 함수를 종료합니다")  # 1~ 99번째 함수의 호출을 종료한다

recursive_function(1)  # 어떤 결과가 출력될지 생각해보자
```

재귀함수 종료하기 이어서...


* 컴퓨터 내부에서 재귀 함수의 수행은 스택 자료구조를 이용한다 <- *무슨 의미일까?*
* 함수를 계속 호출했을 때 가장 마지막에 호출한 함수가 먼저 수행을 끝내야 그 앞의 함수 호출이 종료되기 때문이다
* 컴퓨터의 구조 측면에서 보자면 연속해서 호출되는 함수는 메인 메모리의 스택 공간에 적재되므로 재귀 함수는 스택 자료구조와 같다는 말은 틀린 말이 아니다


✅ 재귀 함수는 내부적으로 스택 자료구조와 동일하다는 것만 기억하자. 따라서 스텍 자료구조를 활용해야 하는 상당수 알고리즘은 재귀 함수를 이용해서 간편하게 구현될 수 있다. DFS가 대표적인 예이다

* 재귀 함수를 이용하는 대표적 예제로는 팩토리얼(Factorial) 문제가 있다
* n! = 1 x 2 x 3 x ... x (n-1) x n을 의미한다
* 팩토리얼 함수는 n이 1 이하가 되었을 때 함수를 종료하는 재귀 함수의 형태로 구현할 수 있다.

```
# 반복적으로 구현한 n!
def factorial_iterative(n):
    result = 1
    # 1부터 n까지의 수를 차례대로 곱하기
    for i in range(1, n+1):
        result *= i
    return result

# 재귀적으로 구현한 n!
def factorial_recursive(n):
    if n <= 1:  # n이 1 이하인 경우 1을 반환
        return 1
    return n * factorial_recursive(n-1)
```

```
# 각각의 방식으로 구현한 n! 출력(n=5)
print("반복적으로 구현: ", factorial_iterative(5))
print("재귀적으로 구현: ", factorial_recursive(5))
```

### 재귀 함수의 장점
*재귀함수의 장점은 무엇일까?*<br>

반복문 대신 재귀함수를 사용할 경우, 코드가 더 간결한 것을 알 수 있다.(재귀 함수가 수학의 점화식(재귀식)을 그대로 소스코드로 옮겼기 때문이다)

* 👨‍🔬 점화식: 수학에서 점화식이란 특정한 함수를 자신보다 더 작은 변수에 대한 함수와의 관계로 표현한 것을 의미한다.
    * 1) n이 0혹은 1일 때: factorial(n) = 1
    * 2) n이 1보다 클 떄: factorial(n) = n x factorial(n-1)

```
# 스스로 작성해보자

def factorial_recursive(n):
    if n <=1:
        return 1
    return n * factorial_recursive(n-1)
```

```
print(factorial_recursive(5))
```
