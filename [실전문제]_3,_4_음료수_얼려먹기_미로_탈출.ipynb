```
n, m = map(int, input().split())

# graph 생성
graph = []
for i in range(n):
  graph.append(list(map(int, input())))


# 상, 하, 좌, 우 탐색
# dx = [-1, 1, 0, 0]
# dy = [0, 0, -1, 1]

# dfs 탐색 시작!
def dfs(x, y):
  # 주어진 범위를 벗어나는 경우에는 즉시 종료
  if x <= -1 or y <=-1 or x >=n or y >=m:
    return False
  # 현재 노드를 아직 방문하지 않았다면
  if graph[x][y] == 0:
    # 해당 노드를 방문 처리
    graph[x][y] = 1
    # 상, 하, 좌, 우의 위치를 모두 재귀적으로 호출한다
    dfs(x-1, y)
    dfs(x+1, y)
    dfs(x, y-1)
    dfs(x, y+1)
    return True  # 탐색 후 return True
  return False

# 이제 모든 위치에 대해서 dfs를 수행
result = 0
for i in range(n):
  for j in range(m):
    # 현재 위치에서 DFS 수행
    if dfs(i, j) == True:
      result += 1

    
print("정답: ", result)
```
